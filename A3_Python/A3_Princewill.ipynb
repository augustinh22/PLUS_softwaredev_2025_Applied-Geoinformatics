```python
#import ee and folium
import folium
import pandas as pd
```

```python
#working on how to get the different bike station,create a simple map, and add points to each of the location using folium in python
#Getting the bike locations and saving to a data is done by Pandas, pandas allows us to import data from various file formats such as CSV,JSON etc
```

```python
location = "https://data.smartdublin.ie/dataset/33ec9fe2-4957-4e9a-ab55-c5e917c7a9ab/resource/2dec86ed-76ed-47a3-ae28-646db5c5b965/download/dublin.csv"
```

```python
bikes_station_locations = pd.read_csv(location)
```

```python
bikes_station_locations
```

```python
#The bike stations has been imported.
#The Bikes_Stations_locations has a total of 110 data
#Now i need to keep the latitude and longitude and the Name of the Stations. 
#The last two columns will allow me to map the location and the Name will be use to give each location pin a name
```

```python
bikes_station_locations = bikes_station_locations[["Latitude", "Longitude", "Name",]]
```

```python
#The next step is to create the map of the stations using the location parameters.
#I zoomed to 15
```

```python
map = folium.Map(location=[53.348800, -6.281637],zoom_start=15, control_scale=True)
```

```python
map
```

```python
for index, location_info in bikes_station_locations.iterrows():
    folium.Marker([location_info["Latitude"], location_info["Longitude"]], popup=location_info["Name"]).add_to(map)
```
